```python
#lets work on the nan condition
#here we are considering the datasets of csv file named Bengaluru_House_Data
```

```python
dot=pd.read_csv("Bengaluru_House_Data.csv")
dot.head()
```

```python
dot1=dot.copy()
dot1.columns=list("123456789")
dot1.columns
```

```python
dot.isnull()
```

```python
print(dot.isnull().sum().sum())
dot.isnull().sum()
```

```python
#so we counted the number of nan positions 
#we need to properly deal with these otherwise it will bring flaws to the classifier
#if we are training a dataframe into an algorithm to get a certain prediction
```

```python
dot2=dot.fillna("no value")
#using the above command, the dataframe consisting of the nan cell
#will be automatically filled with the desired value provided by the user
#irrespective of the datatype present in each columns
dot2
```

```python
dot2.dtypes
#so we can see, most of the columns have got a datatype of object
#as object contains all datatypes
```

```python
dtaa=dot.copy()
dtaa.head()
```

```python
dtaa.isnull().sum()
#the society and balcony columns are filled with nan conditions
#we have to deal them
```

```python
dtaa["society"].fillna(dtaa["society"].mean,inplace=True)
dtaa.isnull().sum()
```

```python
dtaa["balcony"].unique()
```

```python
dtaa["balcony"].value_counts()
```

```python
k=dtaa['balcony'].value_counts().idxmax()
#using this method we can get the mode of a column(balcony) in a dataframe
k
```

```python
dtaa["balcony"].fillna(k,inplace=True)
dtaa.isnull().sum()
```

```python
#so we have dealt with the balcony column
#lets drop tupel containing the nan values based on the columns of size and location 
dtaa.dropna(subset=["size","location"],inplace=True)
```

```python
dtaa.isnull().sum()
```

```python
dtaa["bath"].fillna(dtaa["bath"].mean,inplace=True)
dtaa.isnull().sum().sum()
```

```python
#we can see that there is no nan values left in the dataframe
#and we have dealt with proper values such that the predictions made by these
#values dont affect the answers.
```
